```python
## DATA OVERVIEW

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline 

# Load Data
tracks = pd.read_csv('/content/fma-rock-vs-hiphop.csv')
echonest_metrics = pd.read_json('/content/echonest-metrics.json')
echo_tracks = echonest_metrics.merge(tracks[['track_id', 'genre_top']], on='track_id')
echo_tracks.head()
```

```python
## EDA VISUALIZATIONS

sns.set_style('darkgrid')

# Feature correlation matrix
echo_tracks2 = pd.get_dummies(echo_tracks)
plt.figure(figsize=(10,10))
sns.heatmap(echo_tracks2.corr(), cmap='Purples', annot=True)
plt.xticks(rotation=45)
print(echo_tracks.info(), '\n', echo_tracks.head())

# Plotting Function
def plots(df, col):
  if df[col].dtypes == 'float64':
    plt.figure(figsize=(10,10))
    g = sns.boxplot('genre_top', col, data=df)
  plt.xlabel('Genre')
  plt.ylabel(col)
  plt.title('Relational EDA Plots')

columns = echo_tracks.columns
columns = ['acousticness', 'energy', 'danceability', 'liveness', 'instrumentalness', 'speechiness', 'tempo', 'valence']
for col in columns:
  plots(echo_tracks, col)
```

```python
## MODEL BUILDING

# Split the dataset
features = echo_tracks.drop(['track_id', 'genre_top'], axis=1).values
labels = echo_tracks['genre_top'].values
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, random_state=10)

# Scale features before PCA evaluation
scaler = StandardScaler()
scaled_train_features = scaler.fit_transform(train_features)
scaled_test_features = scaler.transform(test_features)

# Explained variance ratios from PCA using all features
pca = PCA()
pca.fit(scaled_train_features)
exp_variance = pca.explained_variance_ratio_
fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(10,10)
ax1.bar(range(pca.n_components_), exp_variance)
ax1.set_xlabel('Principal Component #')

# Cumulative explained variance
cum_exp_variance = np.cumsum(exp_variance)
ax2.plot(cum_exp_variance)
ax2.axhline(y=0.85, linestyle='--')

# Perform PCA with the chosen number of components and project data onto components
pca = PCA(n_components=6, random_state=10)
train_pca = pca.fit_transform(scaled_train_features)
test_pca = pca.transform(scaled_test_features)
```

```python
## MODEL COMPARISON

# Setting up pipelines
tree_pipe = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=6)), 
                      ("tree", DecisionTreeClassifier(random_state=10))])
logreg_pipe = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=6)), 
                        ("logreg", LogisticRegression(random_state=10))])
kf = KFold(n_splits=10)

# Train our models using KFold cv
tree_score = cross_val_score(tree_pipe, features, labels, cv=kf)
logit_score = cross_val_score(logreg_pipe, features, labels, cv=kf)

# Model Selection
print("Decision Tree:", np.mean(tree_score), "Logistic Regression:", np.mean(logit_score))
```

```python
## MODEL EVALUATION

# Train our logistic regression
logreg = LogisticRegression(random_state=10)
logreg.fit(train_features, train_labels)
pred_labels_logit = logreg.predict(test_features)

# Evaluation Metrics
class_rep_log = classification_report(test_labels, pred_labels_logit)
print("Logistic Regression: \n", class_rep_log)
print(confusion_matrix(test_labels, pred_labels_logit), '= Confusion Matrix')
```

```python
## MODEL REVALUATION

# Subset a balanced proportion of data points
rap_only = echo_tracks.loc[echo_tracks['genre_top'] == 'Hip-Hop']
rock_only = echo_tracks.loc[echo_tracks['genre_top'] == 'Rock']
rock_only = rock_only.sample(rap_only.shape[0], random_state=10)
rock_rap_balanced = pd.concat([rock_only, rap_only])

# The features, labels, and pca projection are created for the balanced dataframe
features2 = rock_rap_balanced.drop(['genre_top', 'track_id'], axis=1) 
labels2 = rock_rap_balanced['genre_top']

# Redefine the train and test set with the pca_projection from the balanced data
train_features2, test_features2, train_labels2, test_labels2 = train_test_split(features2, labels2, random_state=10)
scaled_train_features2 = scaler.fit_transform(train_features2)
scaled_test_features2 = scaler.transform(test_features2)

# Re-training PCA
pca2 = PCA(n_components=6, random_state=10)
train_pca2 = pca2.fit_transform(scaler.fit_transform(train_features2))
test_pca2 = pca2.transform(scaler.transform(test_features2))

# Train our logistic regression on the balanced data
logreg2 = LogisticRegression(random_state=10)
logreg2.fit(train_pca2, train_labels2)
pred_labels_logit2 = logreg2.predict(test_pca2)

# Evaluation Metrics
class_rep_log2 = classification_report(test_labels2, pred_labels_logit2)
print("Logistic Regression: \n", class_rep_log2)
print(confusion_matrix(test_labels2, pred_labels_logit2), '= Confusion Matrix')
```
